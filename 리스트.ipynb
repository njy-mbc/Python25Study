```python
# 문자열 관련 함수들

a = "hobby"
print(a.count('b'))

a = "Python is the best choice"
print(a.find('b'))
print(a.find('choice'))

a = "Life is too short"
a.index('t')

# 문자열 삽입
print(",".join('acdc'))

# 소문자 -> 대문자
a = "hi"
print(a.upper())

# 대문자 -> 소문자
a = "HI"
print(a.lower())

# 왼쪽 공백 지우기
a = "     a     "
print(a.lstrip())

# 오른쪽 공백 지우기
a = "   b   "
print(a.strip())

# 문자열 바꾸기

a = "Life is too short"
a.replace("Life", "Your dick")

# 문자열 나누기
a = "Life is too short"
a.split()

b = "a,c,d,c"
b.split(',')

# 특정 문자열 시작
s = "Life is too short"
print(s.startswith("Life"))

# 특정 문자열 끝
print(s.endswith("short"))
print(s.endswith("Life"))
```
